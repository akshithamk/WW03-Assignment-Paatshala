```python
data
```

```python
plt.figure(figsize=(8,8))
plt.hist(data['SL'],color='green',rwidth=0.9)
plt.title('Histogram of Iris dataset',fontsize=16)
```

```python
data[['SL','SW']].plot.kde()
```

```python
#scatterplot
```

```python
plt.figure(figsize=(8,8))
plt.scatter(data['SL'],data['PL'],s=15,c='red',marker='^')
plt.title('Plot of SL vs PL',fontsize=16)
plt.xlabel('SL')
plt.ylabel('PL')
```

```python
#pair plots
```

```python
sns.pairplot(data)
```

```python
# Box plot
```

```python
plt.boxplot(data['PW'])
plt.title('Box plot of PW')
plt.xlabel('PW')
```

```python
#violin plot
```

```python
sns.violinplot(x='PW',y='SW',data=data)
```
